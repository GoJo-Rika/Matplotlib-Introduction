## Matplotlib: pyplot.plot format strings and other options

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x, y, 'b+--')
```

```python
# same effect as 'b+' format strings
plt.plot(x, y, color='blue', marker='+', linestyle='')
```

### Markersize

```python
plt.plot(x, y, color='blue', marker='+', linestyle='', markersize=20)
```

### Alpha property to control transparency of the line chart

```python
plt.plot(x, y, 'g<', alpha=0.5) # alpha can be specified on a scale 0 t0 1
```
