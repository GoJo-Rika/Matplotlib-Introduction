## Saving chart to a file using savefig

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ['Home Rent', 'Food', 'Phone/Interbet Bill', 'Car', 'Utilities']
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%1.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig('piechart.jpg', bbox_inches='tight', pad_inches=1, transparent=True)
```
